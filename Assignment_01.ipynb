```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #import libary
```

```python
df = pd.read_csv('Height_Weight_Data.csv') #set an object to read file
```

```python
df.head()
```

# Scatter Visualization

```python
plt.scatter(df['Height'],df['Weight'])
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height and Weight')
plt.savefig('1.pdf')
```

# spilit Data Set

```python
x = df[['Height']]
y = df[['Weight']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = .25)
```

# model creation

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(xtrain,xtrain)
```

```python
df['predicted Height'] = reg.predict(x)
```

```python
df
```

```python
pred = reg.predict(xtest)
```

```python
pred
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(ytest,pred)
```

```python
mse
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(ytest,pred)
```

```python
mae
```

# Best fit line

```python
plt.scatter(xtest,ytest)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height and Weight')
plt.plot(xtest.squeeze(),reg.predict(xtest),marker = 'o', color = 'Blue')
```

```python
plt.scatter(xtrain,ytrain)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height and Weight')
plt.plot(xtrain.squeeze(),reg.predict(xtrain),marker = 'o', color = 'Blue')
```
